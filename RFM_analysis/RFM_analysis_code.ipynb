```python
df['결제방법'].unique() # 결제방법이 분할결제가 있어 이를 파악
```

```python
df['결제방법'].replace('적 + 포', '적립금', inplace=True)
df['결제방법'].replace('신 + 포', '신용카드', inplace=True)
df['결제방법'].replace('신 + 적', '신용카드', inplace=True)
df['결제방법'].replace('신 + 적 + 포', '신용카드', inplace=True)
df['결제방법'].replace('신 + 웰컴', '신용카드', inplace=True)
df['결제방법'].replace('후 + 포', '후불', inplace=True)
df['결제방법'].replace('후 + 적 + 포', '후불', inplace=True)
df['결제방법'].replace('후 + 적', '후불', inplace=True)
df['결제방법'].replace('정 + 적', '정기결제', inplace=True)
df['결제방법'].replace('정 + 적 + 포', '정기결제', inplace=True)
df['결제방법'].replace('정 + 포', '정기결제', inplace=True)
df['결제방법'].replace('현 + 포', '현금간편결제', inplace=True)
df['결제방법'].replace('현 + 적', '현금간편결제', inplace=True)
df['결제방법'].replace('현 + 적 + 포', '현금간편결제', inplace=True)
df['결제방법'].replace('-', 'x', inplace=True)
df['결제방법'].replace(' + 포', '포인트', inplace=True)

#분할 결제시 맨 앞 결제 방법으로 전환하는 작업
```

```python
df['결제방법'].fillna('x', inplace=True) #nan 값 없애기
```

```python
df = df.drop(df[df['결제방법']== 'x'].index) # nan 값이 드간 행 없애기
```

```python
df['결제방법'].unique() # nan 값 없애고 결제방법 확인
```

```python
grouped_df_m3 = df.groupby('결제방법')  # 결제방법 별 데이터 묶기
```

```python
m3=grouped_df_m3.sum() # 묶은 데이터끼리 합
```

```python
m3['결제방법']=m3.index # m3 인덱스 값으로 컬럼 지정
```

```python
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
# 한글 출력
sns.barplot(data = m3,
             x = '결제방법',y = '판매금액')
plt.xticks(rotation=90)     # 글씨 세로로 배열
plt .show()

# 결제방법마다 총 합산한 결제금액
```
