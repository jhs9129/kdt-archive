```python
from langchain.prompts import PromptTemplate
from langchain.llms import OpenAI

llm = OpenAI(temperature=0)
prompt = PromptTemplate(
    input_variables=['product'],
    template="{product}을 만드는 회사의 이름을 무엇으로 하면 좋을까?"
)

#LLMChain을 사용해서 연결해줌
from langchain.chains import LLMChain

chain = LLMChain(llm=llm, prompt=prompt)

#실행
chain.run("굉장한 파괴력의 무기")

#출력: '\n\n주도하는 무기(Leading Weapon)이라는 이름이 어울리는 것 같습니다.
```

```python
file = open('D:/final_project/EBS/EBS중학뉴런수학2(하)_손석민/01강이등변삼각형(1)_EBS중학뉴런수학2(하).smi', 'r',encoding = 'utf-8')
wholefile = file.readlines()
```

```python
import pandas as pd
```

```python
text_line = []
for line in wholefile:
  line = line.strip()
  mod_line = re.sub(r"\<[^()]*\>", "", line) # regular expression: remove <> part
  korean_line = re.sub(r"[^ㄱ-ㅣ가-힣\s]","",mod_line) 
  if korean_line != '': #mod_line에 내용이 없을 때 즉, ''인 경우는 패스
    text_line.append(korean_line)
text_pd = pd.DataFrame(text_line,columns=['Line']) #pandas dataframe으로 변환 
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate

api_key = ""
chat_model = ChatOpenAI(openai_api_key=api_key)

my_template = input("GPT에게 질문할 때 요구사항을 입력해주세요.")

script = []

for sentence_list in text_sentence_list:    

    text = sentence_list

    template = f"{my_template}" + " 질문: {question}"

    prompt = PromptTemplate.from_template(template)

    script.append(chat_model.predict(prompt.format(question = text)))
```
