# Line & Scatter

참고) pip install plotly

## Basic 1

```python
# 1. 모듈 불러오기
import plotly.graph_objects as go
import numpy as np

#2. Data지정하기
random_x = np.random.randint(0, 100, 20)
random_y = random_x + 100

#3. 차트 정의하기
trace = go.Scatter(x = random_x, y = random_y, mode = 'lines + markers', name = 'Random Variable')
#mode = 'markers' / 'lines' / 'lines + markers'

#4. 레이아웃 
layout = go.Layout(title = 'Basic')

#5. 도화지 만들기
fig = go.Figure(data = trace, layout = layout)

#6. 도화지 실행
fig.show()
```

## Basic 2

```python
import plotly.graph_objects as go
import numpy as np

random_x = np.random.randint(0, 100, 20)
random_y = np.random.randint(0, 100, 20)

trace = go.Scatter( x = random_x, y = random_y, mode = 'markers',
                  marker = dict(size = 20, symbol = 'pentagon'))

layout = go.Layout(title = 'Random Variables',
                  xaxis = {'title': 'Random X'},
                  yaxis = {'title': 'Random Y'})

fig = go.Figure(data = trace, layout = layout)

fig.show()

```

## Basic 3 _ 두개 이상의 차트를 한 화면(fig)에 그리는 방법

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

num = 50

random_var1 = np.random.randint(0, 100, num)
random_var2 = np.random.randint(0, 100, num)
random_var3 = np.random.randint(0, 100, num)

index = pd.date_range('2022-01-01', periods = num)

df = pd.DataFrame([random_var1, random_var2, random_var3]).T
df.index = index
df.columns = ['Random_Var1', 'Random_Var2', 'Random_Var3']

trace1 = go.Scatter(x = df.index, y = df['Random_Var1'], mode = 'lines', name = df.columns[0])
trace2 = go.Scatter(x = df.index, y = df['Random_Var2'], mode = 'lines', name = df.columns[1])
trace3 = go.Scatter(x = df.index, y = df['Random_Var3'], mode = 'lines', name = df.columns[2])

data = [trace1, trace2, trace3]

layout = go.Layout(title = 'Random Variables')

fig = go.Figure(data = data, layout = layout)

fig.show()
```

# Bar Chart

## Basic 1

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
#months = pd.date_range('2022-01-01', periods = 12)
#months = ['ABCDEGSDFASDFASDFASDFASDFASDFASDF', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

random_1 = np.random.randint(0, 100, 12)
random_2 = np.random.randint(0, 500, 12)

trace1 = go.Bar(x = months, y = random_1, name = 'random_1', marker_color = 'indianred')
trace2 = go.Bar(x = months, y = random_2, name = 'random_2', marker_color = 'lightsalmon')

layout = go.Layout(xaxis_tickangle = -45)

data = [trace1, trace2]

fig = go.Figure(data = data, layout = layout)

fig.show()
```

## Basic 2 _ 특정한 부분에 색만 다르게 하고 싶다면?

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

random_1 = np.random.randint(0, 100, 12)
random_2 = np.random.randint(0, 500, 12)

colors1 = ['indianred'] * 12
colors1[4] = 'crimson'
colors2 = ['lightsalmon'] * 12
colors2[4] = 'LightSkyBlue'

trace1 = go.Bar(x = months, y = random_1, name = 'random_1', marker_color = colors1)
trace2 = go.Bar(x = months, y = random_2, name = 'random_2', marker_color = colors2)

layout = go.Layout(xaxis_tickangle = -45)

data = [trace1, trace2]

fig = go.Figure(data = data, layout = layout)

fig.show()
```

## Basic 3 _ 텍스트 나타내기

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

random_1 = np.random.randint(0, 100, 12)
random_2 = np.random.randint(0, 500, 12)

colors1 = ['indianred'] * 12
colors1[4] = 'crimson'
colors2 = ['lightsalmon'] * 12
colors2[4] = 'LightSkyBlue'

texttemplate = ['{}'.format(i) for i in random_1]

trace1 = go.Bar(x = months, y = random_1, name = 'random_1', marker_color = colors1, texttemplate = texttemplate,
               textposition = 'inside')

#textposition = "inside" / "outside"

trace2 = go.Bar(x = months, y = random_2, name = 'random_2', marker_color = colors2)

layout = go.Layout(xaxis_tickangle = -45)

data = [trace1, trace2]

fig = go.Figure(data = data, layout = layout)

fig.show()
```

# Pie Chart

## Basic 1

```python
import plotly.graph_objects as go

labels = ['Apple', 'Banana', 'Tomato', 'Strawberry', 'Melon']
values = [4500, 3000, 2500, 3500, 5000]

trace = go.Pie(labels = labels, values = values, textinfo = 'label+value')

fig = go.Figure(data = trace)

fig.show()
```

## Basic 2 _ 특정 부분 강조

```python
import plotly.graph_objects as go

labels = ['Apple', 'Banana', 'Tomato', 'Strawberry', 'Melon']
values = [4500, 3000, 2500, 3500, 5000]

pull = [0, 0, 0.1, 0, 0]

trace = go.Pie(labels = labels, values = values, textinfo = 'label+value', pull = pull)

fig = go.Figure(data = trace)

fig.show()
```

# Box Plot

## Basic 1

```python
import plotly.graph_objects as go
import numpy as np

random_1 = np.random.randint(0, 100, 20)
random_2 = np.random.randint(0, 100, 20)
random_3 = np.random.randint(0, 100, 20)

trace1 = go.Box(y = random_1, name = 'random_1', fillcolor = 'rgb(1, 100, 156)')
trace2 = go.Box(y = random_2, name = 'random_2', boxpoints = 'all')
trace3 = go.Box(y = random_3, name = 'random_3')

data = [trace1, trace2, trace3]

fig = go.Figure(data = data)

fig.show()
```

# Histogram

## Basic 1

```python
import plotly.graph_objects as go
import numpy as np

random_1 = np.random.randint(0, 1000, 200)

trace = go.Histogram(x = random_1, xbins = dict(size = 10))

fig = go.Figure(data = trace)

fig.show()
```

## Basic 2 _ X범위 조절

```python
import plotly.graph_objects as go
import numpy as np

random_1 = np.random.normal(0, 1, 10000)

trace = go.Histogram(x = random_1, xbins=dict(start = -4, end = 4, size = 0.01))

fig = go.Figure(data = trace)

fig.show()



```

## Basic 3 _ overlay & opacity

```python
import plotly.graph_objects as go
import numpy as np

random_1 = np.random.randn(500)
random_2 = np.random.randn(500) + 1

trace1 = go.Histogram(x = random_1)
trace2 = go.Histogram(x = random_2)

fig = go.Figure(data = [trace1, trace2])

fig.update_layout(barmode = 'overlay')
fig.update_traces(opacity = 0.5)

fig.show()
```

# HeatMap

## Basic

```python
import plotly.graph_objects as go
import numpy as np

random_1 = np.random.randint(-80, 80, 5)
random_2 = np.random.randint(-80, 80, 5)
random_3 = np.random.randint(-80, 80, 5)

x=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
y=['Morning', 'Afternoon', 'Evening']
z = [random_1, random_2, random_3]

trace = go.Heatmap(z = z, y = y, x = x)

fig = go.Figure(data = trace)

fig.show()
```
